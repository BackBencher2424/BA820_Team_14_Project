**The Question 1: Do programming languages naturally cluster into distinct groups based on shared technical features (syntax patterns, file types) and community signals (GitHub activity, Wikipedia view)?**

**Motivation**: Our EDA showed Github_repo_stars and number_of_jobs being extremely right-skewed, with a long tail of thousands of languages that get overlooked when you just look at the obvious top performers. This motivated us to see if all of these langauges are successful or just share some features.

**Why It Matters**: If a college is building a bootcamp or learning platform curriculum, identifying these hidden language families could open up niche markets. Instead of competing head-to-head in saturated spaces like Python and Java, targeted programs around clusters of related languages that share learning curves and community needs could be developed.

**The Surprising Part**: It would be striking if older languages (pre-1980) ended up in the same clusters as modern web languages, purely based on their GitHub engagement patterns. That would suggest community adoption dynamics matter more than when a language was actually created.

**Question 2: Can we segment programming languages into distinct "Market Archetypes": specifically "Hype Driven" (high GitHub stars/subscribers, low job counts) vs. "Silent" (low social hype, high job counts and crude user estimates)?**

**Motivation:** Our EDA showed that github_repo_stars and number_of_jobs are both extremely right-skewed, but they don't always overlap. Some languages dominate the conversation on GitHub for being popular, while others,are actually running production systems and paying people's salaries. The disconnect is striking.

**Why It Matters:** This can act as a risk-assessment tool for CTOs and curriculum leads for separating whats trendy and whats actually in use and show up in job markets with less competition. That's valuable for schools trying to prepare students for real employment rather than just what looks good on a resume.

**The Surprising Part:** It would be striking if corporate-backed languages from tech titans frequently fell into the "Hype-Driven" cluster, suggesting that massive company branding generates more social "noise" than actual industrial deployment compared to grassroots languages.

**The Question 3: Can we detect "Ghost Languages" technologies with significant cultural visibility (high Wikipedia views and rankings) but virtually no real employment ecosystem?**

**Motivation:** Our data shows the median job count is zero, yet language rankings vary dramatically, which is very weird. There's a whole category of languages that look important on paper—they're ranked, they're discussed—but there's no actual work there. That statistical anomaly is worth understanding.

**Decision Impact:** This becomes a practical filter for career counselors. They can spot "Famous-in-Theory" languages that have zero actual job market, protecting students from spending months learning something that won't lead anywhere. It's the difference between being known and being useful.

**The Surprising Part:** Discovering a language in the top 10% globally for Wikipedia visibility that has no recorded jobs and no active development community—a technology that's somehow academically prominent but completely economically dormant.

**The Question 4: Among the 49 technical metadata features we have (syntax patterns, file types, naming conventions), which ones actually predict how long a language survives and remains relevant in the long run?**

**Motivation:** While working with 49 columns, many of them are correlated or sparse and there's a lot of noise to cut through. This brought out the question: what's the minimal set of technical characteristics that actually matter for a language's staying power?

##**Question 1: Do Bob Ross paintings exhibit latent “visual styles” driven more by color palette structure than by the specific landscape elements depicted?**

##What result would be surprising

It would be surprising if the structure of the color palette did not contribute much additional differentiation beyond the landscape elements, especially considering the large differences in the combinations of colors seen in the paintings. Particularly, if paintings with very different color combinations tended to cluster together simply due to similarities in elements, this would be counterintuitive to the strong sense that the colors play a large role in the style.



##Why this question is worth answering

The question of interest is whether the style of the paintings is represented by the way the scene is colored versus the actual elements that are painted. This is a unique data set due to the richness of the binary data for both colors and elements.

##**Question 2 : Do paintings from earlier vs later episodes differ more in composition structure than in subject matter?**

###What outcome would be surprising

It would be surprising if the composition and subject matter of the early and late episodes were indistinguishable, considering the long time span and the variation in the number of elements used in each painting.



###Why this question is worth answering

The data set contains information on the order of episodes and also provides detailed information on the indicators of the elements used, making it possible to differentiate between structural and content evolution.

##**Question 3: Are there distinct painting clusters defined by contrast intensity (light vs dark color mixes), independent of season order?**

##What would be surprising

It would be surprising if the contrast-based groups were closely aligned with the order of the seasons, implying that the strength of contrast varied monotonically over time rather than being a stylistic device employed throughout the series.



##Why this question is worth answering

The EDA results indicate considerable variability in the mix of light and dark colors in the paintings, and this variability is found across a range of seasons.

##**Question 4: Are there recurring “element bundles” that appear together far more often than expected given their individual frequencies?**

##What outcome would be surprising

It would be surprising if element co-occurrence were largely random with no stable bundles emerging, despite strong individual element frequencies being evident in the EDA.



##Why this question is worth answering

The binary nature of the data allows for the direct analysis of co-occurrence patterns across hundreds of paintings, providing insight into whether Bob Ross used compositional templates or individual element choices.
